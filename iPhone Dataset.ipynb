```python
df['Ram'] = df['Ram'].str.extract('(\d+)').astype(int) #Convert the Ram column from object to integer
```

### Create dataframe for the top ten rated IPhones in India

```python
data = df.sort_values(by=['Star Rating'],ascending=False).head(10)
```

```python
data.head()
```

```python
def Barplot(column):
    plt.figure(figsize=(16,12))
    sns.set_style("whitegrid")
    sns.barplot(x=column, y='Product Name', data=data.sort_values(by=[column], ascending = False))
    plt.title(f'Products with the highest {column}', fontsize=18)
    plt.xlabel(column,fontsize=16)
    plt.ylabel('Product Name', fontsize=16)
    plt.xticks(fontsize=14, rotation=45)
    plt.yticks(fontsize=14)
```

```python
Barplot('Number Of Ratings')
```

APPLE IPhone 8 Plus (Gold, 64 GB) is the most rated phone in India.

```python
Barplot('Number Of Reviews')
```

APPLE IPhone 8 Plus (Gold, 64 GB) is also the most reviewed phone

```python
Barplot('Ram')
```

Although APPLE IPhone 8 Plus (Gold, 64 GB) only has 2 GB Ram is stills the most populat IPhone in India.

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='Number Of Ratings', y= 'Sale Price', data= df, size=df["Discount Percentage"], 
                sizes=(100,500), alpha=0.5)
plt.title('Relation Between Sale Price and Number Of Rating', fontsize=18)
plt.xlabel('Sale Price',fontsize=16)
plt.ylabel('Number Of Ratings', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(prop={'size': 18})
plt.show();
```

The negative relation between Sale Price and Number Of Rating means that low price phones are sold more in India.

```python
plt.figure(figsize=(10,8))
sns.scatterplot(y='Number Of Ratings', x= "Discount Percentage", data= df, size=df["Sale Price"], 
                sizes=(100,500), alpha = 0.5)
plt.title('Relation Between Discount Percentage and Number Of Rating', fontsize=18)
plt.xlabel('Discount Percentage',fontsize=16)
plt.ylabel("Number Of Ratings", fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(prop={'size': 18})
plt.show();
```

The positive relation between Discount Percentage and Number Of Rating means that phones of high discount percentage can be sold easly.
